### Data Collection

### Data Formatting

```python
pd.set_option('future.no_silent_downcasting', True)
#convert "FLOODS" column values into numerical (1 for "YES", 2 for "NO")
data["FLOODS"].replace(["YES", "NO"], [1, 0], inplace = True)
```

### Data Cleaning

```python
#remove the "SUBDIVISION" column
data.drop("SUBDIVISION", axis = 1, inplace = True)
```

### Define Features and Target

```python
#Feature variables (features) include all columns except "FLOODS"
features = data.drop("FLOODS", axis = 1)

#Target variable (target) is the "FLOODS" columns
target = data["FLOODS"]
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split

#split the data into training (e.g. 80% training and 20% testing)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 42)

#"X_train" and "y_train" are your training data (features and target)
#"X_test" and "y_test" are your testing data (features and target)
```

###Model Training

```python
from sklearn.linear_model import LogisticRegression

#create a Logistic Regression model
logistic_model = LogisticRegression(max_iter = 1000, random_state = 42)

#train the model based on the training data
# Convert y_train to numeric type before fitting
logistic_model.fit(X_train, y_train.astype(int))
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Create a KNN model(you can specify the number of neighbors, e.g., n_neighbors = 5)
knn_model = KNeighborsClassifier(n_neighbors = 5)

#train the KNN model on the training data
knn_model.fit(X_train, y_train.astype(int))
```

#### Training the DecisionTree Model


```python
from sklearn.tree import DecisionTreeClassifier
# Initialize the decision tree classifier model
decision_tree_model = DecisionTreeClassifier(random_state=42)
# Train the model on the training data
decision_tree_model.fit(X_train, y_train.astype(int))
```

###Model Evaluation

#### Evaluation of LogisticRegression model

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
# Predict using the trained Logistic Regression model
logistic_predictions = logistic_model.predict(X_test)
# Calculate accuracy
logistic_accuracy = accuracy_score(y_test.astype(int), logistic_predictions)
# Generate a classification report
logistic_report = classification_report(y_test.astype(int), logistic_predictions)
# Create a confusion matrix
logistic_confusion_matrix = confusion_matrix(y_test.astype(int), logistic_predictions)
# Calculate ROC-AUC score
logistic_roc_auc = roc_auc_score(y_test.astype(int), logistic_predictions)
# Print evaluation results
print("Logistic Regression Model Evaluation:")
print(f"Accuracy: {logistic_accuracy:.2f}")
print("Classification Report:")
print(logistic_report)
print("Confusion Matrix:")
print(logistic_confusion_matrix)
print(f"ROC-AUC Score: {logistic_roc_auc:.2f}")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
# Predict using the trained KNN model
knn_predictions = knn_model.predict(X_test)
# Calculate accuracy
knn_accuracy = accuracy_score(y_test.astype(int), knn_predictions)
# Generate a classification report
knn_report = classification_report(y_test.astype(int), knn_predictions)
# Create a confusion matrix
knn_confusion_matrix = confusion_matrix(y_test.astype(int), knn_predictions)
# Calculate ROC-AUC score
knn_roc_auc = roc_auc_score(y_test.astype(int), knn_predictions)
# Print evaluation results
print("K-Nearest Neighbors (KNN) Model Evaluation:")
print(f"Accuracy: {knn_accuracy:.2f}")
print("Classification Report:")
print(knn_report)
print("Confusion Matrix:")
print(knn_confusion_matrix)
print(f"ROC-AUC Score: {knn_roc_auc:.2f}")
```

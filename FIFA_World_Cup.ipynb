```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/uc?id=1YPpTVGSwfs7zETBxurz77ZHUSA5_-vIt'
```

```python
wc_data = pd.read_csv(url)
```

```python
wc_data.head()
```

```python
fifa = wc_data.iloc[:, : 78]
```

```python
fifa.head()
```

```python
fifa.info()
```

```python
top10_age = fifa.sort_values(by = ["age"], ascending = False).head(10)
top10_age.head(10)
```

```python
plt.figure(figsize = (9, 9))
sns.barplot(data=top10_age, x='short_name', y='age')
plt.xlabel("Shortname")
plt.ylabel("Age")
plt.title("Top 10 players according to age", size = 30, color = "green")
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels
plt.show()
```

```python
top10_rep = fifa.sort_values(by = ["international_reputation"], ascending = False).head(10)
top10_rep.head(10)
```

```python
top10_rep.info()
```

```python
plt.figure(figsize=(18,6))
plt.plot(top10_rep['short_name'],top10_rep['potential'],color='lime'
          ,label='Potential',marker='d',markersize=13,markeredgecolor='red')
plt.title('Potential of the top 10 players',size=30,color='green')
plt.xlabel('Names',size=24,color='red')
plt.ylabel('Potential',size=24,color='red')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

This file (gen9_pokemon_stats.csv) contains the Types, Stats and Weaknesses of all 103 Pokémon. Each record is a Pokémon.


*   Pokemon - Name of the Pokémon [can be used as a Primary Key    for
other files] -Type 1 - First type of the Pokémon -Type 2 - Second type of the Pokémon, can be empty

*   HP - Base HP stat of the Pokémon

*   Attack - Base Attack stat of the Pokémon

*   Defense - Base Defense stat of the Pokémon

*   Special Attack - Base Special Attack stat of the Pokémon

*   Special Defense - Base Special Defense stat of the Pokémon


*   Speed - Base Speed stat of the Pokémon


*   Weakness-Type - The damage multiplier applied to the Pokémon when hit by a move of this type. For example, 0.5 means the Pokémon would take halve damage, and 2 would mean the Pokémon takes double damage.













```python
url = 'https://drive.google.com/uc?id=1E-ZWPgtjiAyT_315q21CG4Mvkwcy_d2Z'
pokemon_data = pd.read_csv(url)
pokemon_data.head()
```

```python
pokemon_data.columns
```

```python
plt.figure(figsize = (14, 18))
sns.countplot(data = pokemon_data, x = "Type 1")
plt.title("Number of pokemons of each type", size = 20)
plt.xlabel("Type of pokemons")
plt.ylabel("Count")
plt.show()
```

```python
for col in ['Attack', 'Defense', 'HP', 'Special Attack', 'Special Defense', 'Speed']:
    pokemon_data[col] = pd.to_numeric(pokemon_data[col], errors='coerce')
df = pokemon_data.groupby(by='Type 1').mean(numeric_only=True)[['Attack','Defense','HP','Special Attack', 'Special Defense', 'Speed']]
```

```python
df.head()
```

```python
plt.figure(figsize = (10,12))
sns.barplot(x = df.index, y = df['HP'], palette='viridis')
plt.xlabel("Types of pokemons")
plt.ylabel("HP")
plt.title("Average HP of different pokemon types")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df.plot(kind = "bar", figsize = (22, 8))
plt.xticks(rotation = "horizontal")
plt.xlabel("Type of pokemon")
plt.ylabel("Average")
plt.show()
```

```python
weakness_colms= [ 'Weakness-Normal','Weakness-Fire', 'Weakness-Water', 'Weakness-Electric',
       'Weakness-Grass', 'Weakness-Ice', 'Weakness-Fighting',
       'Weakness-Poison', 'Weakness-Ground', 'Weakness-Flying',
       'Weakness-Psychic', 'Weakness-Bug', 'Weakness-Rock', 'Weakness-Ghost',
       'Weakness-Dragon', 'Weakness-Dark', 'Weakness-Steel', 'Weakness-Fairy']
```

```python
for col in weakness_colms:
    pokemon_data[col] = pd.to_numeric(pokemon_data[col], errors='coerce')
type_comparison = pokemon_data.groupby(by = "Type 1").mean(numeric_only=True)[weakness_colms]
```

```python
type_comparison
```

```python
plt.figure(figsize=(14, 9))
sns.heatmap(data=type_comparison,annot=True)
```

```python
import seaborn as sns
print(sns.get_dataset_names())
```

```python
for i in sns.get_dataset_names():
  print(i)
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
iris.head()
```

```python
sns.scatterplot(x = "sepal_length", y = "petal_length", data = iris)
```

```python
sns.scatterplot(x = iris.sepal_length, y = iris.petal_length)
```

```python
sns.displot(iris["sepal_length"])
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
```

```python
sns.histplot(x = "total_bill", data = tips)
```

```python
tips["smoker"].value_counts()
```

```python
tips["size"].value_counts()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "smoker", data = tips)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "size", data = tips)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", style = "size", data = tips)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", style = "time", hue = "size", data = tips)
```

```python
tips.tail()
```

```python
sns.catplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.catplot(x = "smoker", y = "tip", data = tips)
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip)
```

#Supervised Machine Learning


```python
data = [
    ("red", "small", "cherry"),
    ("yellow", "medium", "lemon"),
    ("orange", "medium", "orange"),
    ("green", "medium", "pear")

]
```

```python
def guess_fruit(color, size):
  for item in data:
    if item[0] == color and item[1] == size:
      return item[2]
  return "unkown"
```

```python
color = input("Enter the colour : ")
size = input("Enter the size (small, medium, large) : ")
print(guess_fruit(color, size))
```

#Unsupervised Machine Learning

```python
datas = [
    ("red", "small"),
    ("yellow", "medium"),
    ("orange", "medium"),
    ("green", "medium")
]
```

```python
color_to_number = {'red': 0, 'yellow': 1, 'orange': 2,"green" : 3}
numeric_data = [[color_to_number[color], 'small' if size == 'small' else 'medium'] for color, size, _ in data]
```

```python
numeric_data
```

```python
clusters = {"small" : [], "medium" : []}

#group fruits based on sizes
for fruits in numeric_data:
  clusters[fruits[1]].append(fruits[0])
```

```python
#prediction cluster
new_fruit = ["green", "large"]
nf_num = color_to_number[new_fruit[0]]

prediction = None

if nf_num in clusters["small"]:
  prediction = "small"
elif nf_num in clusters["medium"]:
  prediction = "medium"
```

```python
print("Prediction cluster : ", prediction)
```

```python
Activities = ["School", "Study", "Play", "Sleep", "Others"]
Time_spent = [6, 4, 1.5, 8, 4.5]
plt.figure(figsize = (5, 2))

# Add labels and title
plt.xlabel("Activities")
plt.ylabel("Time spent (in hours)")
plt.title("Time spent by a child during a school day")

#for plotting bar graph
plt.bar(Activities, Time_spent)
```

```python
Activities = ["School", "Study", "Play", "Sleep", "Others"]
Time_spent = [6, 4, 1.5, 8, 4.5]
plt.figure(figsize = (5, 2))

# Add labels and title
plt.xlabel("Activities")
plt.ylabel("Time spent (in hours)")
plt.title("Time spent by a child during a school day")

#for plotting bar graph
plt.bar(Activities, Time_spent, color = "#32a8a8", width = 0.2)
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales = [15000, 18000, 20500, 22000, 19500, 25000, 28000, 24500, 23000, 26000, 30000, 32500]

plt.figure(figsize = (10, 5))

plt.xlabel("Months")
plt.ylabel("Sales")

plt.title("Sales trend over the past 12 months")

plt.bar(months, sales, color = "purple", width = 0.7)
plt.xticks(rotation = 45) # Use plt.xticks to rotate labels


plt.show()
```

```python
categories = ['Housing', 'Utilities', 'Transportation', 'Groceries', 'Entertainment', 'Healthcare', 'Education', 'Miscellaneous']
expenses = [1200, 300, 400, 500, 200, 150, 250, 100]

plt.figure(figsize = (10, 5))

plt.xlabel("Categories")
plt.ylabel("Expenses")

plt.title("Distribution of expenses across categories")

plt.bar(categories, expenses, color = "navy", width = 0.75)
plt.xticks(rotation = 35)

plt.show()
```

# Histogram


```python
import numpy as np
marks = np.random.randint(0, 100, 150)
marks
```

```python
plt.xlabel("marks")
plt.ylabel("no. of students")
plt.title("no. of students vs marks")

#histogram
plt.hist(marks)
#add gridlines
plt.grid(True)
```

```python
plt.figure(figsize = (5, 2))

labels = "S1", "S2", "S3"
sectiones = [70, 30, 100]

plt.pie(sectiones, labels = labels)

plt.title("Pie Chart Example")
plt.show()
```

```python
labels = "S1", "S2", "S3"
sectiones = [70, 30, 100]
colors = ["c", "g", "y"]

plt.pie(sectiones, labels = labels, colors = colors, startangle = 90, explode = (0, 0.1, 0.5), autopct = "%1.4f%%")
plt.legend(title = "three sectiones")
plt.axis("equal")
plt.title("Pie Chart Example")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/uc?id=1Ld-ez8xqx-ntpE7JgBZpSsECrrAOkRSR'
```

```python
titanic_data = pd.read_csv(url)
```

```python
titanic_data.head()
```

```python
titanic_data.info()
```

```python
titanic_data.drop("Cabin", axis = 1, inplace = True)
```

```python
titanic_data.dropna(inplace = True)
titanic_data.reset_index(drop = True)
```

```python
titanic_data.describe()
```

```python
survival_count = titanic_data["Survived"].value_counts()
survival_count
```

```python
plt.figure(figsize = (20, 8))
plt.pie(survival_count, labels = ["Not Survived", "Survived"], startangle = 58, autopct = "%1.2f%%")
plt.title("Number of Passanger survived and not survived", size = 20)
plt.show()
```

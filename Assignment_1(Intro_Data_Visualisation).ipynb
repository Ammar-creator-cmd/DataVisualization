```python
import matplotlib.pyplot as plt
```

```python
sports_name = ["Cricket", "Football", "Basketball", "Volleyball"]
total_student_number = [45, 50, 37, 15]
```

```python
plt.figure(figsize = (10, 10))
plt.bar(x = sports_name, height = [int(x) for x in total_student_number])
plt.xlabel("Type of sports")
plt.ylabel("Total of students")
plt.title("Favourite sports by students")
plt.show()
```

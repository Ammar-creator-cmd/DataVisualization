```python
import matplotlib.pyplot as plt

#some data points (cartesius coordinates)
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 7]
```

```python
plt.plot(x, y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#add title
plt.title("My first Matplotlib Plot")

#show the plot
plt.show()
```

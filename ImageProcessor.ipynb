```python
#function to process an image
def process_image(input_path, output_path):
  #load image from the input path
  image = cv2.imread(input_path)
  #check if the image was loaded successfully(optional tho)
  if image is None:
    print("Error : Unable to load image.")
    return

  #convert the image to grayscale
  grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  #save the processed image to the output path
  cv2.imwrite(output_path, grayscale_image)
  print(f"Image processed and saved as {output_path}")

  #display the input image
  plt.figure(figsize= (10, 5))
  plt.subplot(1, 2, 1)
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY))
  plt.title("Input Image")
  plt.axis("off")

  #display output image
  plt.subplot(1, 2, 2)
  plt.imshow(grayscale_image, cmap="gray")
  plt.title("Output Image")
  plt.axis("off")
```

```python
import cv2

# Input and output file paths (replace these with your own image paths in Google Drive)
input_image_path = '/content/drive/MyDrive/dataset/bmw_input_image.jpg'  # Replace with your input image file path
output_image_path = '/content/drive/MyDrive/dataset/bmw_output_image.jpg'  # Replace with your desired output image file path
# Process the image
process_image(input_image_path, output_image_path)
```

```python
#sort the data by followers in descending order to get top influencers
sorted_data = data.sort_values(by = "Followers", ascending = False)

#Extract the top 10 ovr influencers
top_10_ovr = sorted_data.head(10)

#Extract the top 10 Indian influencers
top_10_indian = sorted_data[sorted_data["Country"] == "IN"].head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_10_ovr['Username'], top_10_ovr['Followers'], label='Top 10 Overall')
plt.bar(top_10_indian['Username'], top_10_indian['Followers'], label='Top 10 Indian')
plt.xlabel('Influencer')
plt.ylabel('Number of Followers')
plt.title('Top 10 Instagram Influencers Overall vs. Top 10 Indian Influencers')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Extract the top 10 Spanish influencers
top_5_spanish = sorted_data[sorted_data["Country"] == "ES"].head(5)

plt.figure(figsize = (12, 6))
plt.bar(top_10_ovr ["Username"], top_10_ovr ["Followers"], label = "Top 10 Overall Influencers")
plt.bar(top_5_spanish ["Username"], top_5_spanish ["Followers"], label = "Top 10 Spanish Influencers")
plt.xlabel("Influencers")
plt.ylabel("Numbers of Followers (n*10^8)")
plt.title("Top 10 Overall Influencers vs Top 10 Spanish Influencers")
plt.xticks(rotation = 50, ha = "right")
plt.legend()
plt.tight_layout
plt.show()
```

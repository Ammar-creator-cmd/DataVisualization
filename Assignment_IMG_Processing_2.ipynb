```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.image as mpimg
```

```python
img = mpimg.imread("/content/drive/MyDrive/dataset/f4p2.jpg") #imread is for reading and finding the image
plt.imshow(img) #imshow is to show the image
plt.axis("off") #turns off axis labels and ticks for a cleaner display
plt.show()
```

```python
def process_image(input_path, output_path):
  image = cv2.imread(input_path)
  if image is None:
    print("Error : Unable to load image.")
    return

  grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  cv2.imwrite(output_path, grayscale_image)
  print(f"Image processed and saved as {output_path}")
  #display input image
  plt.figure(figsize= (10, 5))
  plt.subplot(1, 2, 1)
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY))
  plt.title("Input Image")
  plt.axis("off")

  #display output image
  plt.subplot(1, 2, 2)
  plt.imshow(grayscale_image, cmap="gray")
  plt.title("Output Image")
  plt.axis("off")
```

```python
input_image_path = '/content/drive/MyDrive/dataset/f4p2.jpg'
output_image_path = '/content/drive/MyDrive/dataset/f4p2_grayscale_img.jpg'

process_image(input_image_path, output_image_path)
```

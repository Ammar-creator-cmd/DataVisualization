```python
#function to process an image
def process_image(input_path, output_path):
  #load image from the input path
  image = cv2.imread(input_path)
  #check if the image was loaded successfully(optional tho)
  if image is None:
    print("Error : Unable to load image.")
    return

  #convert the image to grayscale
  grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  #save the processed image to the output path
  cv2.imwrite(output_path, grayscale_image)
  print(f"Image processed and saved as {output_path}")

  #display the input image
  plt.figure(figsize= (10, 5))
  plt.subplot(1, 2, 1)
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY))
  plt.title("Input Image")
  plt.axis("off")

  #display output image
  plt.subplot(1, 2, 2)
  plt.imshow(grayscale_image, cmap="gray")
  plt.title("Output Image")
  plt.axis("off")
```

```python
# Function to find the closest color name for a given RGB color
def closest_color(rgb_color):
    color_names = {
        'Red': (255, 0, 0),
        'Green': (0, 255, 0),
        'Blue': (0, 0, 255),
        'Yellow': (255, 255, 0),
        'Cyan': (0, 255, 255),
        'Magenta': (255, 0, 255),
        'Black': (0, 0, 0),
        'White': (255, 255, 255),
        'Gray': (128, 128, 128),
        "Orange" : (255, 127, 0)

    }


# Load the image
image = io.imread('https://m.media-amazon.com/images/I/61PiVOgLenS._AC_UF894,1000_QL80_.jpg')
# Reshape the image to a 2D array of pixels (height x width, 3 color channels)
image = np.array(image, dtype=np.float64) / 255.0
w, h, d = original_shape = tuple(image.shape)
assert d == 3
image_array = np.reshape(image, (w * h, d))
# Number of dominant colors to detect
n_colors = 10

# Fit K-Means clustering to the image data
kmeans = KMeans(n_clusters=n_colors, random_state=42).fit(image_array)
# Get the RGB values of the dominant colors
dominant_colors = kmeans.cluster_centers_
# Get color names for the dominant colors
color_names = [closest_color(dominant_color * 255) for dominant_color in dominant_colors]


# Display the original image and dominant colors
plt.figure(1)
plt.clf()
plt.axis('off')
plt.title('Original Image')
plt.imshow(image)
plt.figure(2)
plt.clf()
plt.axis('off')
plt.title('Dominant Colors')
# Create a bar chart of the dominant colors with color names
for i, (color, name) in enumerate(zip(dominant_colors, color_names)):
    plt.bar(i, color, color=color)
    plt.text(i, -0.1, name, rotation=45)
plt.show()
```

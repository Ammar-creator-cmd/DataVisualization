```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/uc?id=1id___NYaSi1FYR8N_12N-hCFtB2pJJs7'
```

```python
population_data = pd.read_csv(url)
```

```python
population_data.head()
```

```python
population_data.info()
```

```python
top_10 = population_data.sort_values(by = ["2022 Population"], ascending = False).head(10)
top_10
```

```python
plt.figure(figsize = (10, 15))
sns.barplot(data = top_10, x = "Country/Territory", y = "2022 Population" )
plt.xlabel("Country/Territory")
plt.ylabel("2022 Population")
plt.title(("Country Rankings by 2022 Population wolrdwide"))
plt.xticks(rotation = 58)
plt.show()
```

```python
population_series_ind =  top_10.iloc[1, 5 : 12]
```

```python
population_series_ind
```

```python
plt.figure(figsize = (10, 15))
plt.bar(population_series_ind.index, population_series_ind, width = 0.5, color = "#9e52f0")
plt.title('Change in the population of India in multiple of cr', size=30, color='#7C9D96')
plt.xlabel('Year',size=24)
plt.ylabel('Population',size=24)
plt.xticks(rotation = 30)
plt.show()
```

```python
population_series_chn = top_10.iloc[: 2, 5 : 12]
population_series_chn
```

```python
diff = population_series_chn.iloc[0] - population_series_chn.iloc[1]
diff
```

```python
plt.figure(figsize = (20, 20))
plt.bar(diff.index, diff, width = 0.5)
plt.plot(diff.index, diff, color = "blue", marker = "d", markersize = 8, markerfacecolor = "lavender", markeredgecolor = "black")
plt.title("China vs India Population Data")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

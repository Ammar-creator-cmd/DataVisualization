```python
Activities = ["School", "Study", "Play", "Sleep", "Others"]
Time_spent = [6, 4, 1.5, 8.5, 4.5]
import matplotlib.pyplot as plt
plt.figure(figsize = (6, 4))
#add labels and title
plt.xlabel("Activities")
plt.ylabel("Time spent (in hours)")
plt.title("Time spent by a child during a day")
#for plotting bar graph
plt.bar(Activities, Time_spent)
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales = [15000, 18000, 20500, 22000, 19500, 25000, 28000, 24500, 23000, 26000, 30000, 32500]

import matplotlib.pyplot as plt

#create a line plot
plt.bar(months, sales, color = "#ee82ee")

#add labels and title
plt.xlabel("Months")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly Sales Data")

#for setting x lables vertical
plt.xticks(range(len(months)), months, rotation = "vertical")

#rotate the x axis labels for better raeadability
plt.xticks(rotation = 45)

#display plot
plt.show()

```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales = [15000, 18000, 20500, 22000, 19500, 25000, 28000, 24500, 23000, 26000, 30000, 32500]

#create a line plot
plt.plot(months, sales, marker = "o", color = "#ee82ee")
#add lables n title
plt.xlabel("Months")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly Sales Data")

# for setting x vertically
plt.xticks(rotation = 90)

#display plot
plt.show()
```

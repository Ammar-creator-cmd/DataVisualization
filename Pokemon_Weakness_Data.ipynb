This file (gen9_pokemon_stats.csv) contains the Types, Stats and Weaknesses of all 103 Pokémon. Each record is a Pokémon.


*   Pokemon - Name of the Pokémon [can be used as a Primary Key    for
other files] -Type 1 - First type of the Pokémon -Type 2 - Second type of the Pokémon, can be empty

*   HP - Base HP stat of the Pokémon

*   Attack - Base Attack stat of the Pokémon

*   Defense - Base Defense stat of the Pokémon

*   Special Attack - Base Special Attack stat of the Pokémon

*   Special Defense - Base Special Defense stat of the Pokémon


*   Speed - Base Speed stat of the Pokémon


*   Weakness-Type - The damage multiplier applied to the Pokémon when hit by a move of this type. For example, 0.5 means the Pokémon would take halve damage, and 2 would mean the Pokémon takes double damage.
